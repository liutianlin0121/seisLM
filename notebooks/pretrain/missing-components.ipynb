```python

import seisbench.data as sbd
import matplotlib.pyplot as plt
import tqdm

data = sbd.NEIC()

NUM_TRACES_TO_COLLECT = 5

num_std_zero_waveforms = 0
progress_bar = tqdm.tqdm(range(len(data)))

for i in progress_bar:
    waveforms = data.get_waveforms(i)
    std = waveforms.std(axis=1)
    if std.min() == 0:
        num_std_zero_waveforms += 1

    # Calculate the percentage
    if i > 0:  # Avoid division by zero for the first iteration
        percentage = (num_std_zero_waveforms / i) * 100
    else:
        percentage = 0

    # Update the progress bar description with the current count and percentage
    progress_bar.set_description(
      f"Zero STD Waveforms: {num_std_zero_waveforms} ({percentage:.2f}%)")




    # std_zero_waveforms[i] = waveforms
    # if len(std_zero_waveforms) > NUM_TRACES_TO_COLLECT:
      # break

# for idx, waveforms in std_zero_waveforms.items():
#   plt.figure(figsize=(5, 2))
#   plt.plot(waveforms.T)
#   plt.title(f"Trace {idx} in NEIC")
#   plt.show()
```
